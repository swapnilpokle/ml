```python
import requests
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
API_key = '070c6e069ff5a0428695b0b4c0bfbe8b'
countries = ['China','United States','Turkey','Saudi Arabia','Jamaica']

country_name = []
max_temp = []
min_temp = []
humidity = []
windspeed = []

for country in countries:
  url = f'http://api.openweathermap.org/data/2.5/weather?q={country}&APPID={API_key}&units=metric'
  r = requests.get(url)
  data = r.json()
  formatted_json = json.dumps(data,sort_keys = True,indent=4)
  print(formatted_json)
  country_name.append(country)
  max_temp.append(data['main']['temp_max'])
  min_temp.append(data['main']['temp_min'])
  humidity.append(data['main']['humidity'])
  windspeed.append(data['wind']['speed'])

df = pd.DataFrame()
df['Country'] = country_name
df['Max_Temp'] = max_temp
df['Min_Temp'] = min_temp
df['Humidity'] = humidity
df['Windspeed'] = windspeed

df.head()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
iqr = Q3 - Q1
min = Q1 - 1.5*(iqr)
max = Q1 + 1.5*(iqr)
df = df[(df['price']>min) & (df['price']<max)]
df.head()
```

```python
df.columns
```

```python
le = LabelEncoder()
df['mainroad'] = le.fit_transform(df['mainroad'])
df['guestroom'] = le.fit_transform(df['guestroom'])
df['basement'] = le.fit_transform(df['basement'])
df['hotwaterheating'] = le.fit_transform(df['hotwaterheating'])
df['airconditioning'] = le.fit_transform(df['airconditioning'])
df['furnishingstatus'] = le.fit_transform(df['furnishingstatus'])
df['prefarea'] = le.fit_transform(df['prefarea'])
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('AQI Data Set.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
New_col = ['Id','Month','PM10','SO2','NOx','PM25','NH3','O3','CO','Benzene','AQI']
df.columns = New_col
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
def line_plot(pollutant,color):
  plt.figure(figsize = (12,6))
  plt.plot(df['Month'],df[pollutant],color=color)
  plt.xlabel('Date')
  plt.ylabel('AQI')
  plt.title(f'{pollutant} Trend Over Time')
  plt.xticks(rotation=45)
  plt.show()
```

```python
line_plot('AQI','blue')
```

```python
line_plot('PM10','crimson')
```

```python
line_plot('PM25','cyan')
```

```python
line_plot('CO','lightgreen')
```

```python
line_plot('SO2','orange')
```

```python
line_plot('NOx','purple')
```

```python
# BAr Plot to compare accross various date

def bar_plot(pollutant,color):
  plt.figure(figsize=(12,6))
  plt.bar(df['Month'],df[pollutant],color=color)
  plt.title(f'Comparison Across Months')
  plt.xlabel('Months')
  plt.ylabel(f'{pollutant} value')
  plt.xticks(rotation=45)
  plt.show()
```

```python
bar_plot('AQI','blue')
bar_plot('PM10','crimson')
bar_plot('PM25','cyan')
bar_plot('CO','lightgreen')
bar_plot('SO2','orange')
bar_plot('NOx','purple')
```

```python
#Box plots to analyze the distribution of AQI values for different pollutant categories
plt.figure(figsize=(12, 6))
df.boxplot(color='blue')
plt.title("Distribution of AQI Values for Pollutants")
plt.show()
```

`PM10` levels are directly proportional to the AQI Level.

```python
#Scatter plots to explore the relationship between AQI values and pollutant levels
plt.figure(figsize=(12, 6))
plt.scatter(df["AQI"], df["PM25"], c="red", label="PM2.5")
plt.scatter(df["AQI"], df["SO2"], c="blue", label="SO2")
plt.scatter(df["AQI"], df["O3"], c="green", label="O3")
plt.title("Scatter Plot: AQI vs. Pollutant Levels")
plt.xlabel("AQI Value")
plt.ylabel("Pollutant Level")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('retail_sales_data.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.drop(["invoice_no", "customer_id", "gender", "age", "payment_method"], axis=1, inplace=True)
df.head()
```

```python
df['Sales'] = df['quantity'] * df['price']
df.head()
```

```python
# Groupby data by region, sales
region_sales = df.groupby('shopping_mall')['Sales'].sum()
region_sales.plot(kind='bar')
plt.title('Sales Distribution According to Regions')
plt.xlabel('Region')
plt.ylabel('Total Sales Amount')
plt.show()
```

```python
print(f"The top-performing region is: {region_sales.idxmax()}")
```

```python
# Stacked Bar Plot
region_cat_sales = df.groupby(['shopping_mall','category'])['Sales'].sum().unstack()
region_cat_sales.plot(kind='bar',stacked=True)
plt.title("Sales Comparison by Region and Product Category")
plt.xlabel("Region")
plt.ylabel("Total Sales Amount")
plt.legend(title="Product Category")
plt.show()
```
